# 栈和队列

## 概述

**容器**是一种数据结构。两类最常用的容器：栈(stack)和队列(queue)

栈和队列用于在计算过程中**保存临时数据**，如果可能生成的数据个数可以事先确定，就可以**设置几个变量临时存储数据**。但如果需要存储的数据个数不能确定，就需要**缓存存储**。

栈和队列是使用最多的**缓存存储结构**

### 数据使用顺序

+ 栈是保证数据后进先出（Last In First Out, LIFO）
+ 队列是保证数据先进先出（First In First Out, FIFO）

## 栈：概念和实现

栈（堆栈）可以基于顺序表和链接表实现，执行入栈和出栈操作的一端为栈顶，另一端为栈底

+ 顺序表实现：表的尾端插入和删除元素都是o(1)操作，尾端作为栈顶。可以基于Python的list实现。
+ 链接表实现：首端插入删除是o(1)操作，作为栈顶。可以基于链接表技术和节点类实现

## 栈的应用

### 算术运算

+ 前缀表达式
+ 中缀表达式
+ 后缀表达式

前缀和后缀表达不需要引入括号，后缀表达式特别适合计算机运算

### 栈和递归

如果在一个定义中引用了被定义的对象本身，这种定义为**递归定义**

在一个数据结构里的某个或几个部分具有与整体同样的结构，这是一种**递归结构**

**递归 = 递推 + 回归** ，递归要有递归定义的出口。

递推和回归的过程需要一个栈（缓存结构）支持递归函数的运行，称为**程序运行栈**

任何一个递归定义的函数都可以通过引入一个栈保存中间结果的方式改写成非递归定义的形式，任何一个包含循环的程序都可改写为一个不包含循环的递归函数。
